## Performance measures across datasets

In addition to the system results (logreg), we include results for three baselines. 

- *Most frequent* always predicts the same label: the one most often observed in the training data.
- *Uniform* is a random baseline which predicts all labels with equal probabilility.
- *Stratified* makes random predictions but with the same overall label proportions as in the training data.

```python
f1 = D[~D.bag_features].groupby(('system', 'dataset')).f1.mean().unstack(0)
f1
```
