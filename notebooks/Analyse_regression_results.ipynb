```python
import json
from pathlib import Path

import pandas as pd

json_file = Path("../results/regression_results.jsonl")

list_of_dicts = [json.loads(line) for line in json_file.open()]
D = pd.DataFrame(list_of_dicts)
D
```

## Performance measures across datasets

Results for MAE and $R^2$ across datasets

```python
mean_absolute_error = D[~D.bag_features].groupby(('system', 'dataset')).mean_absolute_error.mean().unstack(0)
r2 = D[~D.bag_features].groupby(('system', 'dataset')).r2.mean().unstack(0)
```
